```python
import serial
```

```python
import time
```

```python
import datetime as dt
```

```python
import matplotlib.pyplot as plt
```

```python
ser = serial.Serial('COM26', 9600, timeout=1)
```

```python
xs = []
ys = []
```

# 參考資料

* https://pythonforundergradengineers.com/python-arduino-potentiometer.html
* https://learn.sparkfun.com/tutorials/graph-sensor-data-with-python-and-matplotlib/all

```python
data = []

for i in range(0, 10):
        line = ser.readline()
    if line:
        tempC = float(line.decode())
        data.append(tempC)
        print(dt.datetime.now().strftime('%H:%M:%S.%f'), tempC)
        xs.append(dt.datetime.now().strftime('%H:%M:%S.%f'))
        ys.append(tempC)
    time.sleep(2)

print(data)
```

```python
# Draw plot
plt.plot(xs, ys)

# Format plot
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.30)
plt.title('DHT11 Temperature over Time')
plt.ylabel('Temperature (deg C)')

# Draw the graph
plt.show()
```

```python
if ser.isOpen():
    ser.close()
```
